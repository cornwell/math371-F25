```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn 
seaborn.set_style('darkgrid')
```

```python
data =pd.read_csv('../../DataSets/Auto.csv', index_col=0)
```

```python
x = data['weight']
y = data['acceleration']
```

```python
plt.scatter(x,y)
plt.show()
```

```python
tr_ixs = np.random.choice(len(x), int(0.06*len(x)), replace=False)
test_ixs = np.array([i for i in range(len(x)) if (not i in tr_ixs)])
```

```python
coeff_arr = np.polyfit(x.iloc[tr_ixs], y.iloc[tr_ixs], 12)
```

```python
my_poly = np.poly1d(coeff_arr)
my_poly.coefficients
```

```python
xlin = np.linspace(x.min(), x.max(),200)

plt.scatter(x.iloc[tr_ixs], y.iloc[tr_ixs], edgecolors='black')
plt.plot(xlin, my_poly(xlin), c='crimson')
plt.ylim((0,35))
plt.show()
```

```python
t_i = np.random.choice(len(test_ixs), 20)
plt.scatter(x.iloc[t_i], y.iloc[t_i], c='orange', edgecolors='black')
plt.plot(xlin, my_poly(xlin), c='crimson')
plt.ylim((0,35))
plt.show()
```
